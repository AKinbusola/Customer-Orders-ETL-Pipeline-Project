```python
import pandas as pd
import numpy as np
import mysql.connector
import sqlalchemy
```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

mysql_user = "####"
mysql_password = "#####" 
encoded_password = quote_plus(mysql_password)
```

```python
# raw database
raw_db = create_engine(
    f"mysql+mysqlconnector://{mysql_user}:{encoded_password}@127.0.0.1/raw_data"
)
```

```python
df_raw = pd.read_sql("SELECT * FROM orders", con=raw_db)
df_raw
```

```python
df_raw = df_raw.rename(columns = {'ï»¿Customer ID': 'Customer ID', 'Date Order was placed': 'Order Date', 'Total Retail Price for This Order':'Total Retail Price'})

df_raw.head(10)
```

```python
df_raw.columns = df_raw.columns.str.lower()
df_raw.columns = df_raw.columns.str.replace(' ', '_')
df_raw.head()
```

```python
df_raw['customer_id'] =  df_raw['customer_id'].astype(str).str.zfill(10)
df_raw.head()
```

```python
df_raw['customer_id'] = 'C' + df_raw['customer_id'].astype(str)
df_raw['order_id'] = 'ORD' + df_raw['order_id'].astype(str)
df_raw['product_id'] = 'PRD' + df_raw['product_id'].astype(str)

df_raw['customer_status'] = df_raw['customer_status'].str.strip().str.title()

df_raw.head()
```

```python
df_raw['total_profit'] = (df_raw['total_retail_price'] - df_raw['quantity_ordered'] * df_raw['cost_price_per_unit']).round(2)
df_raw.head()
```

```python
df_raw['order_date'] = pd.to_datetime(df_raw['order_date'], format="%d-%b-%y")
df_raw['delivery_date'] = pd.to_datetime(df_raw['delivery_date'], format="%d-%b-%y")
```

```python
df_raw.dtypes
```

```python
# analytics database
analytics_db = create_engine(
    f"mysql+mysqlconnector://{mysql_user}:{encoded_password}@127.0.0.1/analytics_data"
)
```

```python
df_raw.to_sql(
    "orders",
    con=analytics_db,
    if_exists="replace",
    index=False,
    chunksize=1000
)
```
